```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
```

```python
# Create a dataframe with columns named 'x' and 'y' filled with random numbers
df = pd.DataFrame({"x": np.random.normal(size=50), "y": np.random.normal(size=50)})

# Create an ordinary least squares regression on the data and show summary statistics.
# This formula regresses column 'y' on 'x'
ro = sm.OLS.from_formula("y ~ x", df)
res = ro.fit()
print(res.summary())

# Plot the data
sns.regplot('x', 'y', df)
plt.show()
```

```python
DATA_ROOT = 'ALL CSV FILES - 2nd Edition/'

df = pd.read_csv(DATA_ROOT+'Advertising.csv')
df.info()
df.describe()
```

```python
# TV
df_TV= df[["TV","sales"]]

ro = sm.OLS.from_formula("sales ~ TV", df_TV)
res = ro.fit()
print(res.summary())

sns.regplot('TV', 'sales', df_TV)
plt.show()
```

```python
# Radio
df_radio= df[["radio","sales"]]

ro = sm.OLS.from_formula("sales ~ radio", df_radio)
res = ro.fit()
print(res.summary())

sns.regplot('radio', 'sales', df_radio)
plt.show()
```

```python
# Newspaper
df_radio= df[["newspaper","sales"]]

ro = sm.OLS.from_formula("sales ~ newspaper", df_radio)
res = ro.fit()
print(res.summary())

sns.regplot('newspaper', 'sales', df_radio)
plt.show()
```

```python
# The p-values for the three models are very small, except for newspaper, which has a probability of 0.001 that P>|t|
# Comparing the R-square, TV has the greatest value, 0.612, radio has 0.332 and newspaper has the lowest value, 0.052.
# The 95% confidence intervals of the intercepts in the three models are all greater than 0.
```
